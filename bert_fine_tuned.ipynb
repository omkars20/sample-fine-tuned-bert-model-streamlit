```python
from datasets import load_dataset
```

```python
dataset = load_dataset('imdb')
```

```python
# Display the first 5 rows of the training dataset
import pandas as pd
train_df = pd.DataFrame(dataset['train'])
print(train_df.head())
```

```python
# out of 2500 rows take only 500 rows
small_train_dataset = dataset['train'].select(range(500))
print(small_train_dataset)
```

```python
from transformers import BertTokenizer
print("loading the bert tokenizer")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
print(f"the tokenizer is: {tokenizer}")

```

```python
# Tokenize the dataset
print("Tokenizing the dataset with padding and truncation...")
def preprocess_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)

# Preprocess the first 500 rows of the train dataset
small_train_dataset = dataset['train'].select(range(500))
tokenized_train_dataset = small_train_dataset.map(preprocess_function, batched=True)
print("Tokenization complete!")


```

```python
df_tokenized  = pd.DataFrame(tokenized_train_dataset)
```

```python
# model selection

from transformers import BertForSequenceClassification
print("loading bert model for sequence classification....")

model = BertForSequenceClassification.from_pretrained("bert-base-uncased",num_labels = 2)
print("model loaded")
```

```python
# model training
from transformers import TrainingArguments

#set the training argument
print("setting up trainig arguments.....")

training_args = TrainingArguments(
    output_dir='./result',
    num_train_epochs = 2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_dir = './logs',
    logging_steps = 10,
)

print("training the arguments...")
```

```python
# Train the Model
from transformers import Trainer

# set up the trainer

print("setting up the trainer")

trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = tokenized_train_dataset,
)

# start training

print("starting the training")

trainer.train()

print("Training complete....")
```

```python
model.save_pretrained("./saved_model")
tokenizer.save_pretrained("./saved_model")
```
